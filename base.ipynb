```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.style.use('dark_background')
```

```python
def distancia_plano(A,B):
    return np.sqrt((B[0]-A[0])**2 + (B[1]-A[1]))
def distancia_espacio(C,D):
    return np.sqrt((D[0]-C[0])**2 +(D[1])-C[1]**2 + (D[2]-C[2])**2)

A=np.array([1,2])
B=np.array([4,6])

C=np.array([1,2,3])
D=np.array([4,6,8])

distancia_AB=distancia_plano(A,B)
distancia_CD=distancia_espacio(C,D)

print(distancia_AB)
print(distancia_CD)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot([A[0],B[0]],[A[1],B[1]],'ro-')
plt.text(A[0],A[1], 'A',fontsize=12,ha='right')
plt.text(B[0],B[1], 'B',fontsize=12,ha='right')
plt.title('Puntos y Recta en el Plano')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```

```python
ax=plt.subplot(1,2,2,projection='3d')
ax.plot([C[0],D[0]],[C[1],D[1]],[C[2],D[2]],'bo-')
ax.text(C[0],C[1],C[2], 'C',fontsize=12,ha='right')
ax.text(D[0],D[1],D[2], 'D',fontsize=12,ha='right')
ax.set_title('Puntos y Recta en el Plano')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
